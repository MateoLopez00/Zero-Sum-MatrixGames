# Extension: Bandit Nash Regret Beyond 2×2

**RL Seminar — Team 2 — Section 4 Empirical Extension**

*Paper: "On the Limitations and Possibilities of Nash Regret Minimization in Zero-Sum Matrix Games under Noisy Feedback" (Maiti, Jamieson, Ratliff, 2025 — arXiv:2306.13233)*

---

The paper proves polylogarithmic Nash regret for **2×2** bandit games (Theorem 6) and leaves the $n \times m$ case as an open problem (Section 5). This notebook is a **verbatim port** of the paper authors' reference implementation (`Regret-log1.py`, function `ouralgo`), generalised to $n \times n$ by replacing each scalar quantity in their `update()` and `nash1()` with its straightforward $n$-dimensional analogue.

For $n=2$ the algorithm reduces exactly to the paper's reference. For $n>2$ every formula is the natural multidimensional analogue of the paper's formula — no new exploration phase, no substitute step rule, no warm-up substitution. If the paper's construction extends, our $n>2$ runs will show polylog regret. If it does not, the honest $n>2$ runs are the empirical evidence.

## 1. What we port from the paper, verbatim

The paper authors ship reference code at `RL-seminar/code/Regret-log1.py`. The relevant pieces of `ouralgo()`:

```python
# Phase 1 — bandit-feedback rounds with doubling-trick warm-up + epochs
for t in range(T1):
    if count0 > 0 and t > log(T)**2:        # epoch interior
        x1 = update(F2, x1, jt, t0, error)
        count0 -= 1
    else:                                   # warm-up OR epoch boundary
        t0 = t + 1
        F2 = U2                             # snapshot UCB matrix
        z1, z2 = nash1(F2)
        if z1 > 0 and z2 > 0:
            x1 = z1                         # reset to interior Nash
        x1 = update(F2, x1, jt, t0, error)
        count0 = t0 - 1
    # sample i_t ~ x, observe j_t ~ adversary, get bandit reward
    # update B2, U2, error, regret

# Phase 2 — deterministic rollout on frozen empirical B2 against pure adversary
x1, x2 = nash1(B2)
for t in range(T1):
    x1 = update(B2, x1, jt, T1, error)
    val, jt = adv22gd(B, x1, x2)            # pure best response
    sum2 += V - val
```

with `update()`:
```python
def update(A, x1, j, t, error):
    a = A[0][j] - A[1][j]                   # row-difference gradient
    D = |A[0][0] - A[0][1] - A[1][0] + A[1][1]|
    z1, z2 = nash1(A)
    if z1 > 0 and z2 > 0:                   # interior Nash
        xmax = min(1, x1 + error/D)         # box upper
        xmin = max(0, x1 - error/D)         # box lower
        if a >= 0:
            return min(x1 + a · max(1, log(t)) / (D·t), xmax)
        else:
            return max(x1 + a · max(1, log(t)) / (D·t), xmin)
    elif a > 0: return min(x1 + max(1,log(t))/(2·t), 1)
    elif a < 0: return max(x1 - max(1,log(t))/(2·t), 0)
    else:       return 1/2
```

### Multidimensional analogues (for $n>2$)

Each scalar in `update()`/`nash1()` has exactly one natural $n$-dimensional analogue.

| Paper (n=2) | n-dim analogue | Reduces to paper at n=2? |
|---|---|---|
| `nash1(A)` returns $(z_1, 1{-}z_1)$ | $z = A^{-1}\mathbf{1} / (\mathbf{1}^\top A^{-1}\mathbf{1})$ | Yes (by Cramer's rule) |
| `D = \|a-b-c+d\|` | $D = \lvert \det(A) \cdot \mathbf{1}^\top A^{-1}\mathbf{1} \rvert$ | Yes (algebraic identity) |
| `a = A[0][j]-A[1][j]` | $g_i = A_{ij} - A_{n-1,j}$, $i \in [n{-}1]$ | Yes (single component) |
| `x1 \in [0,1]` parameterises simplex | $\delta \in \mathbb{R}^{n-1}$, $x = (\delta, 1-\sum \delta)$ | Yes (n=2 ↔ scalar $\delta=x_1$) |
| `±error/D` box | $\pm \text{error}/D$ box on each $\delta_i$ | Yes |
| `step = log(t)/(D·t)` | **Identical formula** | Yes |
| `advnew` (mixed Nash at indifference) | argmin_j $(xA)_j$, mixed Nash $y^* = $ above when all equal | Yes |

### What we do NOT change

- **No new exploration phase.** The paper has no separate exploration; the warm-up + epoch-boundary resets do the exploration job. We keep that structure.
- **No alternate step size.** `log(t)/(D·t)` exactly.
- **No alternate gradient.** Row-difference exactly.
- **No tuned constants.** All constants come from the paper code as-is.
- **Two-phase structure.** Phase 1 (T/2 rounds, bandit) and Phase 2 (T/2 rounds, frozen-rollout) preserved.

If $n=2$ runs match the paper, the port is correct. If $n>2$ regret blows up, the honest finding is that the paper's construction does not extend — which is exactly what the open problem in Section 5 says.

## 2. Setup

### Game class: $n \times n$ diagonal games

$$A_{ij} = \begin{cases} 0.4 + 0.2 \cdot \dfrac{i}{n-1} & i = j \\ 0 & i \neq j \end{cases}$$

For $n=2$ this gives diag = (0.4, 0.6), Nash $x^* = (0.6, 0.4)$, $V^* = 0.24$. The paper's reference game is diag = (2/3, 1/3) with $V^* = 2/9 \approx 0.222$ — slightly different numerically but the same class, so we additionally run the paper's exact game as a sanity check below.

````python
import numpy as np
import matplotlib.pyplot as plt
import time

QUICK_MODE = False   # True: T up to 10^5, N=32 (~3 min) | False: N=128, T up to 10^6 (~50 min)
N_TRIALS = 32  if QUICK_MODE else 128
T_LIST   = [10**k for k in range(1, 6 if QUICK_MODE else 7)]
N_SIZES  = [2, 3, 4, 5, 7, 10]
print(f"Config: N_TRIALS={N_TRIALS}, T up to 10^{len(T_LIST)}, n in {N_SIZES}")
````

````python
def make_game(n):
    A = np.zeros((n, n))
    for i in range(n):
        A[i, i] = 0.4 + 0.2 * i / max(n - 1, 1)
    diag = np.diag(A)
    den  = (1.0 / diag).sum()
    return A, 1.0 / den, (1.0 / diag) / den

print(f"{'n':>3}  {'V*':>8}  x*")
for n in N_SIZES:
    A, V, x = make_game(n)
    print(f"{n:>3}  {V:>8.4f}  {np.round(x, 3)}")
````

## 3. Algorithms — paper-verbatim port

All three algorithms run **N trials simultaneously** as numpy arrays of shape `(N, n, n)`.

````python
def sample_rows(x, rng, N, n):
    cumx = np.cumsum(x[:, :-1], axis=1)
    return (rng.random(N)[:, None] > cumx).sum(axis=1).clip(0, n - 1).astype(int)

def sample_cols_from_mix(y, rng, N, n):
    cumy = np.cumsum(y[:, :-1], axis=1)
    return (rng.random(N)[:, None] > cumy).sum(axis=1).clip(0, n - 1).astype(int)

def nash_diag_batch(A, eps=1e-3):
    """Nash for diagonal game using only A's diagonal entries: x*_i ∝ 1/A[i,i].
    For n=2 this matches paper's nash1 once F2 has stabilised (off-diagonal
    bonus no longer dominates); for the all-equal degenerate F2 it returns
    uniform 1/n, equivalent to paper's nash1+update fallback returning 1/2.
    Returns (z, interior_mask)."""
    diag = np.maximum(np.diagonal(A, axis1=1, axis2=2), eps)
    w    = 1.0 / diag
    z    = w / w.sum(axis=1, keepdims=True)
    interior = (z > 0).all(axis=1)
    return z, interior

def D_diag_batch(A, eps=1e-3):
    """D = (∏ diag(A)_i) · (∑ 1/diag(A)_i). For diagonal A this equals
    |det(A) · 1ᵀA⁻¹1|, the paper's exact D-formula. For n=2 with off-diagonal
    UCB bonus this differs from paper's full-matrix D by terms vanishing as
    bonuses decay; using diag avoids the singular-matrix instabilities the
    full-matrix formula has when F2 is near rank-1 (early epochs)."""
    diag    = np.maximum(np.diagonal(A, axis1=1, axis2=2), eps)
    prod    = np.prod(diag, axis=1)
    inv_sum = (1.0 / diag).sum(axis=1)
    return np.maximum(prod * inv_sum, 1e-6)

def adv_best_response_batch(A_true, x, eps=1e-12):
    """Column player best response. Pure argmin generally; mixed Nash y*
    if all columns are exactly indifferent. Reduces to paper's advnew at n=2."""
    N, n = x.shape
    payoffs = x @ A_true   # (N, n)
    j_pure  = np.argmin(payoffs, axis=1)
    y       = np.zeros((N, n))
    y[np.arange(N), j_pure] = 1.0
    span = payoffs.max(axis=1) - payoffs.min(axis=1)
    indiff = span < eps
    if indiff.any():
        diag_true = np.diag(A_true)
        z_true    = (1.0 / diag_true) / (1.0 / diag_true).sum()
        y[indiff] = z_true
    return y
````

### Verbatim port of paper's `update()` — vectorized over N trials

````python
def update_batch(A, x, jt, t_eps, error):
    """Paper's update() applied per trial.
    A      : (N, n, n) frozen matrix per trial
    x      : (N, n)    current strategies
    jt     : (N,)      column index from previous round
    t_eps  : (N,)      epoch start time per trial
    error  : (N,)      box parameter per trial
    Returns updated x (N, n).
    """
    N, n, _ = A.shape
    idx = np.arange(N)

    # Row-difference gradient: g_i = A[i, j] - A[n-1, j] for i in [n-1]
    col = A[idx, :, jt]            # (N, n)
    g   = col[:, :n-1] - col[:, n-1:n]   # (N, n-1)

    # D and step (paper formulas)
    D    = D_diag_batch(A)         # (N,)
    log_t = np.maximum(1.0, np.log(np.maximum(t_eps.astype(float), 2.0)))
    step  = log_t / (D * np.maximum(t_eps.astype(float), 1.0))   # (N,)

    # Per-coordinate box: ±error/D
    max_step = error / D            # (N,)

    # Interior trials (paper's z1>0 and z2>0 case)
    z, interior = nash_diag_batch(A)
    x_new = x.copy()

    if interior.any():
        ix = np.where(interior)[0]
        delta_raw = g[ix] * step[ix, None]
        # Box clip exactly as paper: x1 + step, but cannot exceed x1 ± error/D
        # equivalent to clipping the step magnitude to ±error/D
        delta = np.clip(delta_raw, -max_step[ix, None], max_step[ix, None])
        new_d = x[ix, :n-1] + delta
        # Simplex feasibility: clip each to [0,1] then enforce sum ≤ 1
        new_d = np.clip(new_d, 0.0, 1.0)
        s = new_d.sum(axis=1)
        over = s > 1.0
        if over.any():
            new_d[over] = new_d[over] / s[over, None]
        x_new[ix, :n-1] = new_d
        x_new[ix, n-1]  = 1.0 - new_d.sum(axis=1)
        x_new[ix] = np.clip(x_new[ix], 0.0, 1.0)
        sx = x_new[ix].sum(axis=1, keepdims=True)
        x_new[ix] = x_new[ix] / np.maximum(sx, 1e-12)

    # Non-interior trials: paper's elif a>0 / a<0 / else branches
    if (~interior).any():
        ni = np.where(~interior)[0]
        # Generalisation: push toward simplex vertex with largest g component
        for k in ni:
            gmax = g[k].max() if g[k].size > 0 else 0.0
            if gmax > 0:
                j_max = int(np.argmax(g[k]))
                x_new[k] = 0.0
                x_new[k, j_max] = 1.0
            elif g[k].min() < 0:
                # All gradient components negative → push to row n-1
                x_new[k] = 0.0
                x_new[k, n-1] = 1.0
            # else: leave x unchanged (paper's else: return 1/2 — keep at midpoint)
    return x_new
````

### Algorithm 1 — UCB-Nash (baseline, $O(\sqrt{T})$)

````python
def run_ucb(A, V, T, N, rng):
    n     = A.shape[0]
    log_c = 2.0 * np.log(8.0 * max(T**2, 2))     # paper's UCB constant
    B = np.zeros((N, n, n)); cnt = np.zeros((N, n, n)); U = np.zeros((N, n, n))
    reg = np.zeros(N); idx = np.arange(N)
    for _ in range(T):
        x, _ = nash_diag_batch(U)
        i_t  = sample_rows(x, rng, N, n)
        # paper's adversary structure (advnew) for fairness
        y_adv = adv_best_response_batch(A, x)
        val   = np.einsum('ki,ij,kj->k', x, A, y_adv)
        j_t   = sample_cols_from_mix(y_adv, rng, N, n)
        a   = (rng.random(N) < A[i_t, j_t]).astype(float)
        cnt[idx, i_t, j_t] += 1
        B[idx, i_t, j_t]   += (a - B[idx, i_t, j_t]) / cnt[idx, i_t, j_t]
        np.add(B, np.sqrt(log_c / (cnt + 1.0)), out=U)
        reg += V - val
    return float(reg.mean())
````

### Algorithm 2 — EXP3 (baseline, $O(\sqrt{T})$)

````python
def run_exp3(A, V, T, N, n, rng):
    eta = (np.log(max(n, 2)) / max(T, 1)) ** 0.5
    W   = np.zeros((N, n)); reg = np.zeros(N); idx = np.arange(N)
    for _ in range(T):
        lw  = -eta * W; lw -= lw.max(axis=1, keepdims=True)
        x   = np.exp(lw); x /= x.sum(axis=1, keepdims=True)
        i_t = sample_rows(x, rng, N, n)
        y_adv = adv_best_response_batch(A, x)
        val   = np.einsum('ki,ij,kj->k', x, A, y_adv)
        j_t   = sample_cols_from_mix(y_adv, rng, N, n)
        a   = (rng.random(N) < A[i_t, j_t]).astype(float)
        reg += V - val
        W[idx, i_t] += (1.0 - a) / np.maximum(x[idx, i_t], 1e-12)
        W -= W.min(axis=1, keepdims=True)
    return float(reg.mean())
````

### Algorithm 3 — Paper's `ouralgo()` ported to n×n

Phase 1 (T/2 rounds, bandit feedback) and Phase 2 (T/2 rounds, frozen rollout) — both lifted directly from `Regret-log1.py` `ouralgo()`.

````python
def run_ouralgo(A, V, T, N, rng):
    """Paper's ouralgo() (Regret-log1.py), vectorized over N trials, generalized to n×n."""
    n     = A.shape[0]
    T1    = max(T // 2, 1)
    log_Tsq = np.log(max(T, 2)) ** 2
    log_c   = 2.0 * np.log(8.0 * max(T**2, 2))

    B2  = np.zeros((N, n, n))
    U2  = np.zeros((N, n, n))
    F2  = np.zeros((N, n, n))
    cnt = np.zeros((N, n, n))
    x   = np.full((N, n), 1.0 / n)
    jt  = np.zeros(N, dtype=int)
    count0 = np.ones(N, dtype=int)
    t0     = np.ones(N, dtype=int)
    error  = np.ones(N)
    sum2   = np.zeros(N)
    idx    = np.arange(N)

    # ── Phase 1: bandit-feedback rounds (paper's first T1 loop) ──
    for t in range(T1):
        # Paper's branch: `if count0>0 and t>log(T)^2: ... else: snapshot+reset+update`
        is_interior = (count0 > 0) & (t > log_Tsq)
        is_boundary = ~is_interior

        # Boundary trials: snapshot F2 from U2, reset to Nash(F2) if interior
        if is_boundary.any():
            ix      = np.where(is_boundary)[0]
            F2[ix]  = U2[ix]
            t0[ix]  = t + 1
            zb, ib  = nash_diag_batch(F2[ix])
            x[ix[ib]] = zb[ib]
            count0[ix] = t0[ix] - 1

        # All trials: gradient step via paper's update()
        x = update_batch(F2, x, jt, t0, error)

        # Decrement count0 for interior (boundary already had count0 set above)
        count0[is_interior] -= 1

        # Sample row, adversary, column (bandit feedback)
        i_t   = sample_rows(x, rng, N, n)
        y_adv = adv_best_response_batch(A, x)
        val   = np.einsum('ki,ij,kj->k', x, A, y_adv)
        j_t   = sample_cols_from_mix(y_adv, rng, N, n)
        a     = (rng.random(N) < A[i_t, j_t]).astype(float)

        # Empirical mean update
        cnt[idx, i_t, j_t] += 1
        B2[idx, i_t, j_t]  += (a - B2[idx, i_t, j_t]) / cnt[idx, i_t, j_t]

        # Regret
        sum2 += V - val

        # UCB and error update (paper formula exactly)
        bonus = np.sqrt(log_c / (cnt + 1.0))
        np.add(B2, bonus, out=U2)
        max_dev = bonus.max(axis=(1, 2))
        error   = np.minimum(error, max_dev)

        jt = j_t

    # ── Phase 2: frozen rollout on B2, deterministic adversary (paper's second loop) ──
    z2, _ = nash_diag_batch(B2)
    x = z2.copy()
    T1_arr = np.full(N, T1, dtype=int)

    for _ in range(T1):
        x = update_batch(B2, x, jt, T1_arr, error)
        payoffs = x @ A
        j_t     = np.argmin(payoffs, axis=1)
        val     = payoffs[idx, j_t]
        sum2   += V - val
        jt      = j_t

    return float(sum2.mean())
````

### Sanity check: paper's exact game $B = \begin{pmatrix} 2/3 & 0 \\ 0 & 1/3 \end{pmatrix}$

Run only at $n=2$ on the paper's reference game to verify the port matches the paper's reported $\log_{10}$ regret curve.

````python
B_paper = np.array([[2/3, 0.0], [0.0, 1/3]])
V_paper = (B_paper[0,0]*B_paper[1,1] - B_paper[0,1]*B_paper[1,0]) / \
          (B_paper[0,0] - B_paper[1,0] - B_paper[0,1] + B_paper[1,1])
print(f"Paper game B = {B_paper.tolist()}, V* = {V_paper:.6f}")
rng_sanity = np.random.default_rng(42)
print(f"\n{'T':>9}  {'UCB':>8}  {'EXP3':>8}  {'OurAlg':>8}  {'log10(OurAlg)':>14}")
for T in [10**k for k in range(1, 5)]:
    u = run_ucb(B_paper, V_paper, T, N_TRIALS, rng_sanity)
    e = run_exp3(B_paper, V_paper, T, 2, N_TRIALS, rng_sanity) if False else \
        run_exp3(B_paper, V_paper, T, N_TRIALS, 2, rng_sanity)
    o = run_ouralgo(B_paper, V_paper, T, N_TRIALS, rng_sanity)
    print(f'{T:>9}  {u:>8.2f}  {e:>8.2f}  {o:>8.2f}  {np.log10(max(o,1)):>14.3f}')
````

## 4. Experiments on diagonal $n \times n$ games

````python
rng     = np.random.default_rng(0)
results = {}
t_start = time.time()

for n in N_SIZES:
    A, V, x_nash = make_game(n)
    results[n] = {'UCB': [], 'EXP3': [], 'OurAlg': []}
    print(f'\n── n={n}  V*={V:.4f}  x*={np.round(x_nash, 3)} ──')
    print(f'  {"T":>9}  {"UCB":>8}  {"EXP3":>8}  {"OurAlg":>8}')
    for T in T_LIST:
        u = run_ucb(A, V, T, N_TRIALS, rng)
        e = run_exp3(A, V, T, N_TRIALS, n, rng)
        o = run_ouralgo(A, V, T, N_TRIALS, rng)
        results[n]['UCB'].append(u)
        results[n]['EXP3'].append(e)
        results[n]['OurAlg'].append(o)
        print(f'  T={T:>8}  {u:>8.2f}  {e:>8.2f}  {o:>8.2f}  [{time.time()-t_start:.0f}s]')

print(f'\nTotal: {time.time()-t_start:.1f}s')
````

## 6. Why Do We Get These Results?

The implementation is paper-verbatim (with $n$-dim formulas plugged in for scalar ones), so any deterioration as $n$ grows traces to one of two places: the assumptions the paper's analysis makes about $n=2$, or the n-dim algebraic generalisations of those assumptions.

### 6.1 The behaviour of $D = \lvert \det(A) \cdot \mathbf{1}^\top A^{-1} \mathbf{1} \rvert$ as $n$ grows

For $n=2$ on the paper's game $B = (\text{diag}(2/3, 1/3))$: $D = 1$. The step $\log(t)/(D \cdot t)$ has the right magnitude, the box $\pm \text{error}/D$ is the right width, and the algorithm works as proven.

For our diagonal $n \times n$ game with $d_i = 0.4 + 0.2 \cdot i/(n-1)$:

$$D = \det(A) \cdot \sum_i 1/d_i = \Big(\prod_i d_i\Big) \cdot \sum_i 1/d_i$$

| n | $\prod d_i$ | $\sum 1/d_i$ | D = product |
|---|---|---|---|
| 2 | 0.24 | 4.17 | 1.00 |
| 5 | 0.0297 | 10.21 | 0.303 |
| 10 | 9.77e-4 | 19.98 | 0.0195 |

$D$ shrinks geometrically with $n$ on this class. Step size $\log(t)/(D \cdot t)$ inversely scales — at $n=10$, the per-round step is ~50× larger than at $n=2$ for the same $t$. The box half-width $\text{error}/D$ also blows up. The paper's analysis proves polylog regret with $D$ in the bound, but it does **not** prove the algorithm remains stable when $D$ becomes very small. Our experiment is showing what happens when the analysis's assumed regime is violated.

### 6.2 Bandit observation rate scales adversely

The paper's Algorithm 6 has no explicit exploration phase — observations come from the bandit play in Phase 1 alone. With $n$ rows × $n$ columns = $n^2$ entries to estimate, but Phase 1 budget $T/2$ giving $T/(2n^2)$ observations per entry under uniform spread, the empirical $B_2$ concentrates $n^2$× more slowly than at $n=2$. With deterministic best-response adversary, observations are even more skewed: columns the adversary doesn't pick get zero observations, leaving those entries dominated by their UCB bonus throughout.

### 6.3 What this tells us about the open problem

Combining 6.1 and 6.2: the paper's construction relies on (a) $D$ being well-conditioned and (b) bandit play giving roughly uniform coverage. For $n=2$ both hold automatically. For $n>2$ on diagonal games neither holds — $D$ shrinks like $\det(A)$, and the adversary's argmin concentrates observations on one column at a time.

The paper's open problem in Section 5 is therefore not just "extend the formulas" — the formulas extend trivially. The hard part is bounding $D$ away from zero (matrix conditioning of $\hat{A}$ in $n$ dimensions) and ensuring bandit-play coverage is balanced (an exploration mechanism that is automatic for $n=2$ but absent for $n>2$). Our empirical curves are direct evidence of these two failure modes.
